```python
#checking and detecting   outliers using boxplot
numerical_cols = ["adr", "days_in_waiting_list", "lead_time"]
fig, axes = plt.subplots(1, 3, figsize=(12, 6))
for ax, col in zip(axes, numerical_cols):
    sns.boxplot(y=df[col], ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()

```

```python
#Removing outliers using  IQR(Inter Quartile Range)method
numerical_cols = ["adr", "days_in_waiting_list", "lead_time"]
for col in numerical_cols:
  q1=df[col].quantile(0.25)
  q3=df[col].quantile(0.75)
  iqr=q3-q1
  lower_bound=q1-1.5*iqr
  upper_bound=q3+1.5*iqr
  df=df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# Q1. What is the average lead time for bookings that are canceled versus those that are not canceled?

average_lead_time = (
    df.groupby("is_canceled")["lead_time"]
      .mean()
      .reset_index(name="average_lead_time")
)

# Replace 0 and 1 with meaningful labels
average_lead_time["Cancellation_Status"] = (
    average_lead_time["is_canceled"]
    .replace({
        0: "Not Cancelled",
        1: "Cancelled"
    })
)

print(average_lead_time)

# Bar Chart
plt.figure(figsize=(8,5))

sns.barplot(
    data=average_lead_time,
    x="Cancellation_Status",
    y="average_lead_time",
    hue="Cancellation_Status",     # Required for future seaborn versions
    palette="viridis",
    legend=False
)

plt.title("Average Lead Time for Cancelled vs Not Cancelled Bookings")
plt.xlabel("Cancellation Status")
plt.ylabel("Average Lead Time (Days)")

plt.show()
```

```python
# 1. How does the lead time affect the likelihood of booking cancellations?


lead_time_cancel=df.groupby("is_canceled")["lead_time"].mean().reset_index(name="Average_lead_time")
lead_time_cancel["Cancellation Status"] = (
    lead_time_cancel["is_canceled"]
    .replace({0:"Not Cancelled",1:"Cancelled"})
)


print(lead_time_cancel)
print("*"*1000)
# Box Plot showing lead time vs cancellation

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x="is_canceled",
    y="lead_time"
)

plt.xticks([0,1],["Not Cancelled","Cancelled"])
plt.title("Lead Time vs Cancellation")
plt.show()

# Business Insight from Box plot Graph(below)
# Cancelled bookings generally have a significantly higher lead time.
# Guests booking months in advance are more likely to change their travel plans






```

```python
# Q2.What impact do special requests have on booking cancellations?
special_request = (
    df.groupby("total_of_special_requests")["is_canceled"]
      .mean()
      .mul(100)
      .reset_index(name="Cancellation Rate")
)

print(special_request)
print("*"*100)

# Line Chart showing Cancellation Rate vs Special Requests

plt.figure(figsize=(10,6))

sns.lineplot(
    data=special_request,
    x="total_of_special_requests",
    y="Cancellation Rate",
    marker="o"
)

plt.title("Cancellation Rate vs Special Requests")
plt.show()

# Buisness Insights from the below line graph
# Guests with more special requests usually show lower cancellation rates.
#These customers appear more committed to completing their stay.
```

```python
# 3.What trends can be observed in the booking behavior of international guests compared to domestic guests?
# df.info()
df["guest_type"] = df["country"].apply(
    lambda x: "Domestic" if x=="PRT" else "International"
)
guest_behavior = (
    df.groupby("guest_type")
      .agg(
          Total_Bookings=("hotel","count"),
          Cancellation_Rate=("is_canceled","mean"),
          Average_ADR=("adr","mean"),
          Average_Lead_Time=("lead_time","mean")
      )
)

guest_behavior["Cancellation_Rate"]*=100

print(guest_behavior)

#Grouped Bar chart
sns.barplot(
    data=df,
    x="guest_type",
    y="lead_time",
    estimator="mean"
)

# Business Insight
# International guests book earlier.
# International guests pay higher ADR.
# International guests cancel more frequently.
```

```python
# Q4.What is the impact of the day of the week on booking cancellations?
df["arrival_date"] = pd.to_datetime(
    df["arrival_date_day_of_month"].astype(str)
    + "-"
    + df["arrival_date_month"]
    + "-"
    + df["arrival_date_year"].astype(str)
)

df["day_of_week"] = df["arrival_date"].dt.day_name()

day_cancel = (
    df.groupby("day_of_week")["is_canceled"]
      .mean()
      .mul(100)
      .reset_index(name="Cancellation Rate")
)

print(day_cancel)
print("*"*1000)
#BAR CHART below showing the day of week vs cancellation Rate
sns.barplot(
    data=day_cancel,
    x="day_of_week",
    y="Cancellation Rate"
)
# Buisness Insights
# Weekend arrivals have a higher cancellation rate.
# Midweek arrivals are more stable.
```

```python
# 5. What are the most common countries for international guests, and how do their booking patterns compare to domestic guests?
#Internation Guests
international = df[df["country"]!="PRT"]

top_countries = (
    international["country"]
    .value_counts()
    .head(10)
)

print(top_countries)
#Booking Pattern
booking_pattern = (
    df.groupby("guest_type")
      .agg(
          Average_Lead_Time=("lead_time","mean"),
          Average_Length_of_Stay=("length_of_stay","mean"),
          Cancellation_Rate=("is_canceled","mean"),
          Average_ADR=("adr","mean")
      )
)

booking_pattern["Cancellation_Rate"]*=100

print(booking_pattern)

# BAR CHART showing Top 10 International Countries

sns.barplot(
    x=top_countries.values,
    y=top_countries.index
)

#Buisness Insights
# A small number of international countries contribute a large share of
#  foreign bookings, enabling targeted marketing campaigns and personalized guest services.
```
